<h3>Own Account Workers</h3>

```python
df_Own_Account=pd.read_csv("~/data_viz/Project_Employement/data/raw_data/Own-account workers/Own-account Workers.csv")
df_Own_Account.head()
```

```python
def own_account_process(df):
    df=df.drop(columns="symbol")
    df['date']=pd.to_datetime(df['date'])
    df['date'] =df['date'].dt.strftime('%Y')
    return df
```

```python
df_Own_Account_1=own_account_process(df_Own_Account)
df_Own_Account_1.to_csv("~/data_viz/Project_Employement/data/tidy_data/Own-account workers/Own-account Workers.csv")
```

```python
df_Own_Account_Male=pd.read_csv("~/data_viz/Project_Employement/data/raw_data/Own-account workers/Own-account Workers, Male.csv")
df_Own_Account_Male.head()
```

```python
df_Own_Account_Male_1=own_account_process(df_Own_Account_Male)
df_Own_Account_Male_1.to_csv("~/data_viz/Project_Employement/data/tidy_data/Own-account workers/Own-account Workers, Male.csv")
```
