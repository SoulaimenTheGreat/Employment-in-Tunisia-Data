# Violin Plots

```python
import plotly.express as px


fig = px.violin(df_Priv_sec_indicator, y="Indicator value", x="Year", color="Year", box=True, points="all",
          hover_data=df_Priv_sec_indicator.columns,title="")
fig.update_layout(template="plotly_white",
        title={
        'text': "Private sector indicators for tunisia",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
        })
fig.show()
```

```python
df_1960=(grouped.get_group(1960)).sort_values(by=['Indicator value'],ascending=False).head(15)
df_1970=(grouped.get_group(1970)).sort_values(by=['Indicator value'],ascending=False).head(15)
df_1980=(grouped.get_group(1980)).sort_values(by=['Indicator value'],ascending=False).head(15)
df_1990=(grouped.get_group(1990)).sort_values(by=['Indicator value'],ascending=False).head(15)
df_2000=(grouped.get_group(2000)).sort_values(by=['Indicator value'],ascending=False).head(15)
df_2010=(grouped.get_group(2010)).sort_values(by=['Indicator value'],ascending=False).head(15)
df_2019=(grouped.get_group(2019)).sort_values(by=['Indicator value'],ascending=False).head(15)

#fig_pie = px.pie(df_2019, values=df_2019["Indicator value"], names=df_2019["Indicator name"], title='Indicators values 2019', hole=.3)
#fig_pie.show()



fig_pie=go.Figure()



#button layout
fig_pie.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=list([dict(label="1960",
                     method="update",
                     args=[{"visible": [True,False,False,False,False,False,False]},
                           {"title": "Indicators 1960"
                           }]),
                          dict(label="1970",
                     method="update",
                     args=[{"visible": [False,True,False,False,False,False,False]},
                           {"title": "Indicators 1970"
                           }]),
                          dict(label="1980",
                     method="update",
                     args=[{"visible": [False,False,True,False,False,False,False]},
                           {"title": "Indicators 1980"
                           }]),
                           dict(label="1990",
                     method="update",
                     args=[{"visible": [False,False,False,True,False,False,False]},
                           {"title": "Indicators 1990"
                           }]),
                           dict(label="2000",
                     method="update",
                     args=[{"visible": [False,False,False,False,True,False,False]},
                           {"title": "Indicators 2000"
                           }]),
                           dict(label="2010",
                     method="update",
                     args=[{"visible": [False,False,False,False,False,True,False]},
                           {"title": "Indicators 2010"
                           }]),
                dict(label="2019",
                     method="update",
                     args=[{"visible": [False,False,False,False,False,False,True]},
                           {"title": "Anne2019"
                            }])
            ])
        )
    ])


fig_pie.add_trace(
    go.Pie(labels=df_1960["Indicator name"],values=df_1960["Indicator value"],hole=.4))

fig_pie.add_trace(
    go.Pie(labels=df_1970["Indicator name"],values=df_1970["Indicator value"],hole=.4))


fig_pie.add_trace(
    go.Pie(labels=df_1980["Indicator name"],values=df_1980["Indicator value"],hole=.4))



fig_pie.add_trace(
    go.Pie(labels=df_1990["Indicator name"],values=df_1990["Indicator value"],hole=.4))



fig_pie.add_trace(
    go.Pie(labels=df_2000["Indicator name"],values=df_2000["Indicator value"],hole=.4))



fig_pie.add_trace(
    go.Pie(labels=df_2010["Indicator name"],values=df_2010["Indicator value"],hole=.4))


fig_pie.add_trace(
    go.Pie(labels=df_2019["Indicator name"],values=df_2019["Indicator value"],hole=.4))



#lengend layout

fig_pie.update_layout(
    legend=dict(
        x=1,
        y=1,
        traceorder="normal",
        font=dict(
            family="sans-serif",
            size=10,
            color="black"
        ),
      
    ))


fig_pie.show()
```
